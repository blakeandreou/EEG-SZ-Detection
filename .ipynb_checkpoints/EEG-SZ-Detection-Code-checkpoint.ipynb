```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, regularizers
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('mergedTrialData.csv')
df.head()
```

```python
#get demo and compare quick
demo = pd.read_csv('demographic.csv')
demo[['age','education','group']].groupby('group').describe()
```

```python
#isolate N100 data
rawN100 = df.loc[:,df.columns.str.endswith('N100')]
rawN100.insert(loc=0,column='subject',value=df['subject'])
#attach condition
cleanN100 = rawN100.merge(demo[['group','subject']],on='subject').dropna()
cleanN100
```

```python
#logistic model creation
def logisticModel():
    model = tf.keras.models.Sequential([
        #simple logistic model first with 1 nodes see what we're starting with
        layers.Dense(units = 1, activation = 'sigmoid')
    ])
    return model
```

```python
def runModel(data,modelFunc):
    #split data
    training_features, test_features, training_labels, test_labels = train_test_split(data.drop(columns=['subject','group']),data['group'], test_size = 0.1)
    #model creation and compilation
    model = modelFunc()
    model.compile(loss = 'binary_crossentropy',
        optimizer = 'adam',
        metrics = ['Accuracy'])
    model_stats = model.fit(training_features,training_labels,epochs=10,validation_data=(test_features,test_labels))
    return model_stats
```

```python
modelStats = runModel(cleanN100,logisticModel)
```

```python
for metric in ['loss','Accuracy']:
    plt.figure()
    plt.plot(modelStats.history[metric],label='test')
    plt.plot(modelStats.history['val_' + metric],label='validation')
    plt.title(metric.capitalize())
    plt.legend()
    plt.show()
```
